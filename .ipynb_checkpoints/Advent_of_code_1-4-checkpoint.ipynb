```python
def is_strictly_increasing(row):
    values = row.dropna().values
    return all(x < y for x, y in zip(values, values[1:]))

def is_strictly_decreasing(row):
    values = row.dropna().values
    return all(x > y for x, y in zip(values, values[1:]))

def check_monotonicity(row):
    values = row.dropna().values
    if is_strictly_increasing(row):
        return 1
    elif is_strictly_decreasing(row):
        return 1
    else:
        return 0

df['Monotonicity'] = df.iloc[:,0:8].apply(check_monotonicity, axis=1)
df
```

```python
import numpy as np
df = pd.read_csv('input_day_2.txt', sep='\s+', header=None)
def check_row(row):
    values = row.dropna().values
    decreasing = all(x > y for x, y in zip(values, values[1:]))
    increasing = all(x < y for x, y in zip(values, values[1:]))
    monotonicity = 1 if increasing or decreasing else 0
    is_dif = 1 if all(y - 3 <= x <= y + 3 and x != y for x, y in zip(values, values[1:])) else 0
    if monotonicity * is_dif == 1:
        return 1
df['is_dif1'] = df.iloc[:,0:8].apply(check_row, axis=1)
df['is_dif1'].sum()
```

```python
def check_row2(row):
    values = row.dropna().values
    decreasing = all(x > y for x, y in zip(values, values[1:]))
    increasing = all(x < y for x, y in zip(values, values[1:]))
    monotonicity = 1 if increasing or decreasing else 0
    is_dif = 1 if all(y - 3 <= x <= y + 3 and x != y for x, y in zip(values, values[1:])) else 0
    if monotonicity * is_dif == 1:
        return 1
    
    for i in range(len(values)):
        modified_values = np.delete(values, i)
        decreasing = all(x > y for x, y in zip(modified_values, modified_values[1:]))
        increasing = all(x < y for x, y in zip(modified_values, modified_values[1:]))
        monotonicity = 1 if (increasing or decreasing) else 0
        is_dif = 1 if all(abs(y - x) <= 3 and abs(y - x) >= 1 for x, y in zip(modified_values, modified_values[1:])) else 0
        if monotonicity * is_dif == 1:
            return 1
    return 0
    
df['is_dif2'] = df.iloc[:,0:8].apply(check_row2, axis=1)
df['is_dif2'].sum()
```

```python
# 3 PART 2
pattern_mul = r"mul\((0|[1-9]\d{0,2}),(0|[1-9]\d{0,2})\)"
pattern_control = r"don't\(\)|do\(\)"

def extract_mul_values(text):
    blocks = re.split(r"don't\(\)", text)
    
    mul_matches = []
    first_block = blocks[0]
    matches = re.findall(pattern_mul, first_block)
    mul_matches.extend(matches)
    
    for block in blocks[1:]:
        if 'do()' in block:
            # Extract everything after 'do()'
            do_index = block.index('do()') + len('do()')
            block = block[do_index:]  # Everything after 'do()'
        else:
            # If there's no 'do()' in this block, skip it entirely
            continue
        # Extract mul(x,x) values from the current block
        matches = re.findall(pattern_mul, block)
        mul_matches.extend(matches)
    
    return mul_matches

result = extract_mul_values(text)

result = [(int(x), int(y)) for x, y in result]
sum2 = 0
for _ in range(len(result)):
    sum2 += mul(int(result[_][0]), int(result[_][1]))
    
sum2
```

```python
# 3 PART 2 geresnis variantas
import re
def part2(input):
    do = r"do\(\)"
    dont = r"don't\(\)"
    mul = r"mul\((\d+),(\d+)\)"
    enabled = True
    sum = 0
    for x in re.finditer(f'{do}|{dont}|{mul}', input):
        if re.fullmatch(do, x.group()):
            enabled = True
        elif re.fullmatch(dont, x.group()):
            enabled = False
        elif enabled:
            sum += int(x.group(1)) * int(x.group(2))
    return sum
total = part2(text)
total
```

```python
# 4 PART 1 brute force
import numpy as np
text = open("input_day_4.txt", "r").read().splitlines()

text_array = np.array([list(row) for row in text]) 
rows, cols = text_array.shape
xmas = r'(?=(XMAS|SAMX))'
a=0
for row in text:
    a += len(re.findall(r'(?=(XMAS|SAMX))', row))  

for col in range(cols):
    vertical_string = ''.join(text_array[:, col])  
    a += len(re.findall(r'(?=(XMAS|SAMX))', vertical_string))  
    
for _ in range(-rows+1, cols):
    diagonal_tl_to_rb_char = []
    for row in range(rows):
        for col in range(cols):
            if row-col == _:
                diagonal_tl_to_rb_char.append(text_array[row][col])
    diagonal_tl_to_rb = ''.join(diagonal_tl_to_rb_char)
    a += len(re.findall(xmas, diagonal_tl_to_rb))

for _ in range(rows+cols-1):
    diagonal_tr_to_lb_char = []
    for row in range(rows):
        for col in range(cols):
            if col+row == _:
                diagonal_tr_to_lb_char.append(text_array[row][col])
    diagonal_tr_to_lb = ''.join(diagonal_tr_to_lb_char)
    a += len(re.findall(xmas, diagonal_tr_to_lb))

a
```

```python
# 4 PART 1 better

import re
import numpy as np

text = open("input_day_4.txt", "r").read().splitlines()
text_array = np.array([list(row) for row in text]) 
rows, cols = text_array.shape
xmas = r'(?=(XMAS|SAMX))'
a = 0

for row in text:
    a += len(re.findall(xmas, row)) 

for col in range(cols):
    vertical_string = ''.join(text_array[:, col])  
    a += len(re.findall(xmas, vertical_string))  

for offset in range(-rows + 1, cols):
    start_row = max(0, offset)
    start_col = max(0, -offset)
    diagonal_length = min(rows - start_row, cols - start_col)  
    diagonal_characters = [text_array[start_row + i][start_col + i] for i in range(diagonal_length)]  
    diagonal_tl_to_rb = ''.join(diagonal_characters) 
    a += len(re.findall(xmas, diagonal_tl_to_rb))

for offset in range(rows + cols - 1):  
    start_row = max(0, offset - cols + 1)  
    start_col = min(cols - 1, offset)  
    diagonal_length = min(start_col + 1, rows - start_row)  
    diagonal_characters = [text_array[start_row + i][start_col - i] for i in range(diagonal_length)]  
    diagonal_tr_to_lb = ''.join(diagonal_characters)  
    a += len(re.findall(xmas, diagonal_tr_to_lb))  

a
```
