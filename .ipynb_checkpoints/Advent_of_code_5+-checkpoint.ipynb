```python
rule_pairs = set(tuple(pair) for pair in rules)

def is_correctly_sorted(update_list, rule_pairs):
    # Create a dictionary for positions of elements in update_list
    positions = {val: i for i, val in enumerate(update_list)}

    # Check each rule to ensure order is correct if both elements are present
    for a, b in rule_pairs:
        if a in positions and b in positions:
            # a should come before b
            if positions[a] > positions[b]:
                return False
    return True

correct_updates = [upd for upd in updates if is_correctly_sorted(upd, rule_pairs)]
a=0
for _ in range(len(correct_updates)):
    a += correct_updates[_][len(correct_updates[_]) // 2]
a
#statistics.median(correct_updates[_])
```

```python
from collections import defaultdict, deque

def topological_sort(update, rules):
    """Use Kahn's algorithm to create a topological ordering."""
    # Step 1: Build the graph (adjacency list) and in-degree count
    graph = defaultdict(list)
    in_degree = defaultdict(int)
    nodes = set()
    
    for a, b in rules:
        graph[a].append(b)
        in_degree[b] += 1
        nodes.add(a)
        nodes.add(b)
    
    # Step 2: Start with nodes with 0 in-degree
    queue = deque([node for node in nodes if in_degree[node] == 0])
    levels = defaultdict(list)
    level = 1
    
    while queue:
        current_level = list(queue)
        for node in current_level:
            levels[level].append(node)
            queue.popleft()
            for neighbor in graph[node]:
                in_degree[neighbor] -= 1
                if in_degree[neighbor] == 0:
                    queue.append(neighbor)
        level += 1
    
    return levels


def check_update_order(update, rules):
    """Check if the update is in the correct topological order according to the rules."""
    # Step 1: Build a position map for each number in the update
    position = {num: idx for idx, num in enumerate(update)}
    
    # Step 2: Verify that for every rule [a, b], `a` appears before `b` in the update list
    for a, b in rules:
        if a in position and b in position:
            if position[a] > position[b]:  # If a comes after b, it's incorrect
                return 0
    return 1


# Example usage
update = [19, 2, 8, 4, 18, 9, 92, 24]
rules = [[92, 23], [9, 4], [2, 9], [19, 92], [24, 8], [18, 4]]

levels = topological_sort(update, rules)
result = check_update_order(update, rules)
levels, result
```

```python
from pathlib import Path
#Day5_input = open(Path(__file__).parent / "Day5_input.txt").read()

################################
###--------- PART 1 ---------###
################################

associated_rules = {}
for sequence in updates:
    for rule in rules:
        if all(ele in sequence for ele in rule):
            if associated_rules.get(tuple(sequence)) == None:
                associated_rules[tuple(sequence)] = []
            associated_rules[tuple(sequence)].append(rule)

# Topological Sort using Kahn's Algorithm: 

# Count how many times a number must be after another (How many times does it appear as the second number in the rule):
# if no. of times is 0, add it to the sorted list. 
# Remove all rules that have that number as the first number
# Check if more numbers now qualify
# If number remains, list is unsortable.

def kahn_sort(sequence, rules):
    in_degrees = {}
    if len(sequence) == 1:
        return sequence
    for number in sequence:
        in_degrees[number] = 0
        for rule in rules:
            if rule[1] == number:
                in_degrees[number] += 1
        if in_degrees[number] == 0:
            new_seq = [x for x in sequence]
            new_seq.remove(number)
            new_rules = [x for x in rules]
            for rule in rules:
                if rule[0] == number:
                    new_rules.remove(rule)
            sorted_list = kahn_sort(new_seq, new_rules)
            if sorted_list is not False:
                return [number] + sorted_list
    return False

correct_sum = 0
fixed_sum = 0
for sequence in updates:
    result = kahn_sort(sequence, associated_rules[tuple(sequence)])
    if result != sequence:
        fixed_sum += result[len(result) // 2]
    else:
        correct_sum += result[len(result) // 2]

print(f"Sum of correct sequences:{correct_sum}")

################################
###--------- PART 2 ---------###
################################

print(f"Sum of fixed sequences: {fixed_sum}")

```
